```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
```

```python
data = pd.read_csv('/content/drive/MyDrive/uber-raw-data-sep14.csv')
```

```python
data
```

```python
data
```

```python
data['Date/Time'] = data['Date/Time'].map(pd.to_datetime)
```

```python
data
```

```python
data["Day"] = data["Date/Time"].apply(lambda x: x.day())
```

```python
data["Weekday"] = data["Date/Time"].apply(lambda x: x.weekday())
```

```python
data
```

```python
data["Month"] = data["Date/Time"].apply(lambda x: x.month)
```

```python
data['Year'] = data['Date/Time'].apply(lambda x: x.year)
```

```python
data
```

```python
data["Hour"] = data["Date/Time"].apply(lambda x: x.hour)
```

```python
sns.set(rc={'figure.figsize':(20, 12)})
```

```python
sns.distplot(data["Hour"])
```

```python
sns.displot(data['Weekday'])
```

```python
df = data.groupby(["Weekday", "Hour"])
for i in df:
  print(i)
```

```python
df = data.groupby(["Weekday", "Hour"]).apply(lambda x: len(x))
```

```python
df
```

```python
df = df.unstack()
```

```python
df
```

```python
sns.heatmap(df, annot=False)
```
